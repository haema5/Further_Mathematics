### Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
x = np.linspace(-5, 5, 20)

y = 3*x + 1
y2 = -(1/3)*x + 1

plt.plot(x, y)
plt.plot(x, y2)

plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.axes().set_aspect('equal')

plt.show()
```

### 2.3. Напишите код на Python, реализующий построение графиков:

#### 1) окружности

```python
t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t), np.sin(t))

plt.axes().set_aspect('equal')
plt.show()
```

#### 2) эллипса

```python
r1 = 3
r2 = 2

t = np.linspace(0, 2*np.pi, 100)
plt.plot(r1*np.cos(t), r2*np.sin(t))

plt.axes().set_aspect('equal')
plt.show()
```

#### 3) гиперболы

```python
X = np.around(np.arange(-10, 10, 0.1), decimals=1)
Y = []
for x in X:
    if x != 0:
        Y.append(1/x)
    else:
        Y.append(np.nan)
 
plt.plot(X, Y)
plt.show()
```

### 2.5.1. Нарисуйте трехмерный график двух параллельных плоскостей.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

Z = 3*X + 3*Y
ax.plot_surface(X, Y, Z)

D = 30
Z2 = 3*X + 3*Y + D 
ax.plot_surface(X, Y, Z2)

plt.show()
```

### Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = plt.figure(figsize=plt.figaspect(0.4))

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

#Конус:
ax = fig.add_subplot(1, 2, 1, projection='3d')
Z = np.sqrt(X**2 + Y**2)
ax.plot_surface(X, Y, Z)
ax.set_title('Конус')

#Элиптический парабалойд:
ax = fig.add_subplot(1, 2, 2, projection='3d')
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)
ax.set_title('Элиптический парабалойд')

plt.show()
```
